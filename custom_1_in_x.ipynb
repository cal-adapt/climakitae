```python
import pandas as pd
import xarray as xr
from climakitae.core.data_interface import get_data

import numpy as np
```

```python

# Or use specific locations (e.g., major cities)
city_locations = [
    (34.05, -118.25),  # Los Angeles
    (37.77, -122.42),  # San Francisco
    # (38.58, -121.49),  # Sacramento
    # (32.72, -117.16),  # San Diego
    # (36.75, -119.77),  # Fresno
    # (37.34, -121.89),  # San Jose
    # (33.95, -117.40),  # Riverside
    # (37.80, -122.27),  # Oakland
    # (38.55, -121.74),  # Davis
    # (34.42, -119.70),  # Santa Barbara
]
tolerance = 0.01  # degrees

# retrieve min and max temp for these locations 
min_temp_dfs = []
max_temp_dfs = []

min_temp_var = "Minimum air temperature at 2m"
max_temp_var = "Maximum air temperature at 2m"
downscaling = "Statistical"
resolution = "3 km"
timescale = "daily"
approach = "Warming Level"
wls = [1.5, 2.0, 3.0]


```

```python

# Retrieve data for all locations
for lat, lon in city_locations:
    min_temp_dfs.append(
        get_data(
            variable=min_temp_var,
            latitude=(lat - tolerance, lat + tolerance),
            longitude=(lon - tolerance, lon + tolerance),
            downscaling_method=downscaling,
            timescale=timescale,
            approach=approach,
            warming_level=wls,
            resolution=resolution,
        )
    )
    max_temp_dfs.append(
        get_data(
            variable=max_temp_var,
            latitude=(lat - tolerance, lat + tolerance),
            longitude=(lon - tolerance, lon + tolerance),
            downscaling_method=downscaling,
            timescale=timescale,
            approach=approach,
            warming_level=wls,
            resolution=resolution,
        )
    )

```
