### Generate Time-Based Standard Year Profiles

"Precipitation (total)" unit="mm"


```python
stn_file = pkg_resources.resource_filename("climakitae", "data/hadisd_stations.csv")
stn_file = pd.read_csv(stn_file, index_col=[0])
```

```python
stn_file['station'][1:]
```

```python
centered_year = 2015
variable = "Air Temperature at 2m"
units = "degF"

for station in stn_file['station'][1:]:
    print(f"Producting time-based climate profile for {station}, {variable} centered at {centered_year}")
    profile_selections = {
        "variable": variable,
        "resolution": resolution,
        "q": qtile,
        "units": units,
        "no_delta": no_delta,
        # # approach
        "approach": "Time",
        "centered_year": centered_year,
        # # warming level window
        "warming_level_window": window, 
        # # Location options -- uncomment based on your desired location type
        "stations": [station],  # uncomment for a weather station
    }
    profile = get_climate_profile(**profile_selections)
    export_profile_to_csv(profile, **profile_selections)
```
